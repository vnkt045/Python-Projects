```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("cpci.csv")
df
```

```python
plt.ylabel('Year')
plt.xlabel('Per Capita Income')
plt.scatter(df.percapinc,df.year,color='red',marker='+')
```

```python
newdf=df.drop('percapinc',axis='columns')
newdf
```

```python
model=linear_model.LinearRegression()
model.fit(newdf,df.percapinc)
```

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(newdf,df.percapinc)
```
