```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r'E:\Imarticus_R_Project\Imarticus_DSP_29\Pokemon1.csv',header=0,index_col=0)
df
df.shape
```

```python
#Scatter Function
plt.figure(figsize=(20,10))
plt.scatter(data=df,x='Attack',y='Defense')
plt.show()
```

```python
#list of functions used for customizing the graphs
plt.figure(figsize=(20,10))
plt.scatter(data=df,x='Attack',y='Defense',s=150,c='red',marker='^')

#add x and y labels
plt.xlim(0,150)
plt.ylim(0,200)

#add title
plt.title('Relationship between attack and defense')
import numpy as np

#add x and y labels
plt.xlabel('Attacks')
plt.ylabel('Defense')
plt.xticks(np.arange(0,max(df["Attack"])+1,10))
plt.grid(True)

#show plot
plt.savefig("test.png")
plt.show()
```

```python
#scatter plot with regression line
sns.lmplot(data=df,x='Attack',y='Defense')
```

```python
sns.lmplot(data=df,x='Attack', y='Defense',fit_reg=False,hue='Stage',palette="Set1")
```

```python
sns.pairplot(df,x_vars=["Attack","Defense","Speed"],y_vars="Total",kind='reg')
```

```python
#Line plot
x=[10,13,50,70]
y=[10,45,10,70]
plt.plot(x,y,c='purple')

plt.xlim(0,100)
plt.ylim(0,100)
plt.title('A Simple Line Plot')
plt.grid(True)
plt.savefig("test.jpg")
plt.show()
```

```python
#Histogram
plt.hist(x="Attack",data=df,bins=10,color='darkblue')
#plt.hist(x="Defense",data=df,bins=10,color='red')
```

```python
plt.subplot(121)
plt.hist(x="Attack",data=df,bins=10,color='darkblue')
plt.title('Attacks')
plt.subplot(122)
plt.hist(x="Defense",data=df,bins=10,color='darkblue')
plt.title('Defense')
plt.show()
```

```python
objects=df["Type 1"].unique()
#print(objects)

import numpy as np
x_pos=np.arange(len(objects))
y_pos=np.arange(len(objects))
#print(x_pos)
#print(y_pos)

new_df=df.groupby("Type 1").mean()
#print(new_df)

total_attacks=new_df["Attack"]
plt.subplot(121)
plt.barh(y_pos,total_attacks)
plt.yticks(y_pos,objects)
plt.xlabel("Attack Points")
plt.ylabel('Type 1')


plt.subplot(122)
plt.bar(y_pos,total_attacks,alpha=0.7)
plt.xticks(y_pos,objects,rotation=90)
plt.xlabel("Type 1")
plt.ylabel('Attack Points')
```

```python
#pie Charts

import matplotlib.pyplot as plt

#slices will be plotted conter-clockwise:
labels=['SAS','Python','R','Tableau']
sizes=[14.5,45.5,30,10]
explode=[0.1,0,0,0]
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.2f%%',
        colors=["red","green","purple","yellow"],
        shadow=True,startangle=90,counterclock=True)
plt.axis('equal')
plt.show()
```
